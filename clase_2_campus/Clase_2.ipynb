#### **Ejemplo 1 - Gráficos de línea**

````python
# Crear un gráfico de líneas
plt.figure(figsize=(12, 6)) # Ajustar el tamaño del gráfico si se desea
plt.plot(df.sensor0, color='green')
plt.xlabel('Tiempo (seg)') # Etiqueta para el eje x
plt.ylabel('Temperatura (ºC)') # Etiqueta para el eje y
plt.title('Evolución de la temperatura del Sensor 0')

# Mostrar el gráfico
plt.tight_layout() # Ajuste del contenido de gráfico
plt.show()
````

````python
# Crear el gráfico de líneas
plt.figure(figsize=(12, 6))  # Ajustar el tamaño del gráfico si se desea
plt.plot(df.sensor0, color='green')
plt.plot(df.sensor1, color='red')
plt.plot(df.sensor2, color='orange')

plt.legend(['Sensor 0', 'Sensor 1', 'Sensor 2'], loc="center right")
plt.xlabel('Tiempo (seg)')
plt.ylabel('Temperatura (ºC)')
plt.title('Evolución de la temperatura')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
````

#### **Ejemplo 2 - Gráficos de barra**

````python
# Crear el gráfico de barras
plt.figure(figsize=(12, 6))  # Ajustar el tamaño del gráfico si se desea
plt.bar(['Sensor 0', 'Sensor 1', 'Sensor 2'], df.iloc[3, :3], color='skyblue') # Ver abajo sección Pandas Slicing
plt.xticks(rotation = 45)
plt.xlabel('Sensores')
plt.ylabel('Temperatura')
plt.title('Temperatura sensada (Lectura 4)')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
````

#### **Ejemplo 3 - Gráficos de dispersión**

````python
# Crear el gráfico de dispersión
plt.figure(figsize=(12, 6))  # Ajustar el tamaño del gráfico si se desea
plt.scatter(df.sensor0, df.sensor1, color='blue', s=250, alpha=.5)
plt.xlabel('Sensor 0')
plt.ylabel('Sensor 1')
plt.title('Relación lecturas Sensor0 y Sensor1')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
````

## **Otra forma de generar los gráficos**

Matplotlib utiliza una jerarquía básica para manejar los objetos de la interfaz.
Para generar figuras y gráficos podemos hacerlo de forma imperativa u orientada a objetos, ahora utilizaremos la segunda opción que nos permite manipular luego los objetos referenciados.
El objeto figura es el contenedor de nivel superior, se pueden crear figuras con el método plt.figure(), aunque en este curso veremos un método abreviado que nos provee el módulo pyplot.
Una figura puede contener uno o más ejes (lo que matplotlib llama eje es en realidad un par de ejes XY), es un plano donde se puede graficar. 

<img src='https://matplotlib.org/devdocs/_images/anatomy.png' width=600>

Además, Matplotlib ofrece una amplia gama de opciones para personalizar la apariencia de los gráficos. 

A continuación, se presentan algunos ejemplos donde se pueden ajustar:
- Fuentes
- Colores
- Estilos de línea
- Grillas
- Etiquetas

````python
# Creación del gráfico de líneas
# Creación de la figura y un eje
fig, ax = plt.subplots()
# Configurar el alto de la figura (Valor = 5), ancho de la figura (valor = 10),
# color ('lightgrey') y título ('Precipitación mensual para la Provincia de Santa Fe (en milímetros)')
fig.suptitle('Evolución de la temperatura (1 minuto)', fontsize=20 )
fig.set_facecolor('lightgrey')
fig.set_figheight(5)
fig.set_figwidth(10)
# Etiquetas de los ejes x, y
ax.set_xlabel('Tiempo (Segs.)')
ax.set_ylabel('Temperatura (ºC)')
# Etiquetas de las reglas
ax.tick_params(axis='x', labelsize=13, labelrotation=90)
ax.tick_params(axis='y', labelsize=13)
# Grilla horizontal, color gris, linea de puntos, espesor 1.
ax.grid(axis='y', color='grey', linestyle=':', linewidth=1)
# Trazamos con linestyle, color, linewith
ax.plot(df.sensor0[:60], linestyle=':', marker='s', label='Sensor 0')
ax.plot(df.sensor1[:60], marker='o', label='Sensor 1')
ax.plot(df.sensor2[:60], marker='*', label='Sensor 2')
ax.legend()
plt.tight_layout()
plt.show()
````

## **Múltiples gráficos**

Matplotlib permite alojar múltiples gráficos (axes) en una figura, podemos indicar esto mediante el método subplots(). 
En los parámetros pasaremos primero el número de filas y luego el de columnas en la grilla de la figura.
El método sigue devolviendo dos objetos, por un lado la figura y luego los "axes" en una matriz/vector.

Como ejemplo creamos una figura con dos ejes en una columna.

```
fig, ax = plt.subplots(filas, columnas)
```

````python
# Crear un objeto figura (fig) y un objeto eje (ax) con el método subplots().
fig, ax = plt.subplots(2)
````

¿Como creamos dos gráficos en una fila?

````python
# Cambiamos los parámetros pasados al método subplots()
fig, ax = plt.subplots(1, 2)
````

````python
# Dos gráficos en una fila
fig, ax = plt.subplots(1, 2, figsize=(12,6))
fig.suptitle('Sensado de Temperaturas')
ax[0].set_title('Lecturas Sensor 0')
ax[0].plot(df.sensor0, color='green')
ax[0].tick_params(axis='x', labelrotation=45)
ax[1].set_title('Proporción de Muestras (Status)')
ax[1].pie(df.status.value_counts(), labels=['Buen Estado', 'Riesgo de Falla'], autopct='%1.1f%%', startangle=90, shadow=True)
````
